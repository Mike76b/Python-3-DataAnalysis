## Data Analysis
### Example of Data Cleaning // Ejemplo de Limpieza de Datos
***(EN)***    In the following example we can see a picture that shows a massive star formation region* in which I've decide to keep the most shining dots (assuming those are the relevant information in this case) and remove the rest from the picture. To do so:  

***(ES)***    En el siguiente ejemplo podemos ver una imagen de una región donde hay una formación masisva de estrellas*, de la cual he decidido quedarme con los puntos más brillantes (asumiendo que esta es la información más importante en este caso) y eliminar el resto de la imagen. Para ello:

**Thanks to European Space Agency website for sharing this image // Gracias a la web de la Agencia Espacial Europea por compartir esta imagen. https://www.esa.int/spaceinimages/Images/2018/05/Hidden_secrets_of_a_massive_star-formation_region*

```python
import matplotlib.pyplot as plt 
# (EN) The following will show the image without any color changes.
# (ES) Aquí se puede ver la imagen sin modificación alguna de los colores.
photo_d = plt.imread("C:/Users/Mike/Pictures/stars.jpg") 
plt.figure(figsize=(12,12))
plt.imshow(photo_d)
```

***(EN)*** As you can see, there are a lot of colors that make harder to find the shining dots.

***ES)*** Como se puede observar, en la imagen hay muchos colores que dificultan la ubicación de los puntos brillantes.

### Identifying the features // Identificación de las características.
***(EN)*** In the following code I'm looking to get useful information from the file, to know what I might need for its treatment.

***(ES)*** En las siguientes líneas de código estoy buscando información del archivo que podría ser útil para saber qué necesito para su tratamiento.

```python
print("Flags:", photo_d.flags)  # This is useful to know, for example, if it's writeable or not (flag -> WRITEABLE : False). 
                                # Es útil para saber si es editable (indicador -> WRITEABLE : False).

print("Shape:", photo_d.shape)  # This helps to confirm that actually you have one array for each color (Red / Green / Blue) which you can see in the value of last index (3).
                                # Útil para confirmar con el valor del último índice (3) que realmente tienes tres array, una para cada color (Rojo / Verde / Azul).

print("Size:", photo_d.size)    # To know the size / Para conocer el tamaño.

print("Min:", photo_d.min(), "Max:", photo_d.max()) # Useful to confirm that you don´t have wrong outside rage values (0 - 255).
                                                    # Sirve para confirmar que no hay valores erróneos que se salgan del rango (0 - 255).

print("Mean:", photo_d.mean())  # Tells you the mean, that you can use as a point of reference / dice el promedio, que se puede usar como punto de referencia.

print("Type:", type(photo_d))   # To confirm that it's a numpy ndarray. // Para confirmar que es una numpy ndarray.
```

### Making the file writeable and first filter // Cambiando el archivo a modificable y primer filtro
***(EN)*** As you can see, the flag WRITEABLE is setted as False, which means that is not possible to manipulate the values of the arrays. To change that, I'm using the following command.

name_of_object **.setflags(write=1)**

Sometimes, setting the values of one or two colors to zero happens to be all you need get a right view. So here I'm setting to zero all red values and all blue values with the following lines of code.

green_photo_d[ :, :, 0] = 0

green_photo_d[ :, :, 2] = 0

***(ES)*** Como se ve, el indicador WRITEABLE está configurado como Falso, lo que significa que no es posible manipular los valores de las arrays. Para cambiarlo, uso el siguente comando.

nombre_del_objeto **.setflags(write=1)**

A veces, colocar en cero todos los valores de uno o dos de los colores es todo lo que se necesita para ver con claridad lo que se busca. Aquí, estoy colocando en cero todos los valores de rojo y de azul con las siguientes líneas de código.

green_photo_d[ :, :, 0] = 0

green_photo_d[ :, :, 2] = 0

```python
green_photo_d = plt.imread("C:/Users/Mike/Pictures/stars.jpg")
green_photo_d.setflags(write=1)
print("Shape:", green_photo_d.shape)
print("Size:", green_photo_d.size)
print("Min:", green_photo_d.min(), "Max:", green_photo_d.max())
print("Mean:", green_photo_d.mean())
print("Type:", type(green_photo_d))
# Setting the values of red and blue to zero. // Colocando a cero los valores de rojo y azul.
green_photo_d[:, :, 0] = 0 # Between square brackets, 0 refers to red color array // el cero entre corchetes indica que es el array de rojo.
green_photo_d[:, :, 2] = 0 # Between square brackets, 2 refers to blue color array // el cero entre corchetes indica que es el array de azul.
plt.figure(figsize = (12, 12))
plt.imshow(green_photo_d)
```

### Filtering by brightness // Filtro de brillo

***(EN)*** As it seems, filtering by color is not enough, so I'm going to set a values filter to show only the values above 200.

***(ES)*** Como se puede ver, filtrar los colores no fue suficiente (o apropiado), así que voy a filtrar los valores de modo que solo se muestren los valores que están por arriba de 200.

```python
first_filter = photo_d < 200 # <- filtering code. // Código que hace el filtrado.
photo_d.setflags(write=1) 
photo_d[first_filter] = 0    # <- sets to zero the selected range of values. // asigna el valor cero al rango de valores seleccionado.
plt.figure(figsize=(12,12))
plt.imshow(photo_d)
```

***(EN)*** Now is quite easy to see where the shining dots are.

***(ES)*** Ahora es mucho más sencillo ver dónde están los puntos más brillantes.
